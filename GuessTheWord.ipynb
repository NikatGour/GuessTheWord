```python
import random 

name = input("What is your name? ") 

print(f"Good Luck {name}!") 

words = ['rainbow', 'computer', 'science', 'programming', 'absence','academic','kitchen''accompany','knee'
         'accomplish','knife','acknowledge','knock','administration','knowledge','affect','land','analysis',
         'afford','language','landscape','background','balance','large','boundary','capacity',
         'business','campus','capability','challenge','dialogue','python', 'mathematics', 'player', 'condition', 
		'reverse', 'water', 'board', 'geeks'] 

word = random.choice(words) 

print("Guess the characters") 

guesses = '' 

turns = 20

while turns > 0: 
	failed = 0
	for char in word: 
		if char in guesses: 
			print(char) 	
		else: 
			print("*") 
			failed += 1
			
	if failed == 0:  
		print("Congratulations! You Won the Game.") 
		print("The word is:",word) 
		break
	guess = input("guess a character:") 
    
	guesses += guess 
	
	if guess not in word: 
		turns -= 1
		print("Wrong") 
		print("You have", + turns, 'more guesses') 
		if turns == 0: 
			print("You Loose!Better Luck Next Time.") 
```
